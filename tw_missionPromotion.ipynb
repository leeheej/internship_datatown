```python
# ㅇ 데이터 소개
# - (미션형 프로모션) 프로모션 과정에서 주어진 임무를 수행하고 진행 여부에 따라 혜택을 받는 프로모션 입니다.
# - 당사의 고유한 오프라인 인증 기기인 스마트 스탬프 인증 시에 수집하는 정보 입니다.

# ㅇ 활용 분야
# - 유사 프로모션 기획시에 최적화된 방안 도출의 근거자료로 활용
# - 유사 프로모션 집행시에 결과 예측의 근거자료로 활용

# ㅇ 데이터 출처
# - (12CM China) 스탬프 인증 프로모션 사업 수집 정보

df_tw_missionPromotion_202212 = pd.read_csv('./data/TC_TW_MISN_PROMTN_PARTCPTN_CSTMR_INCLN_INFO_20221231.csv')
df_tw_missionPromotion_202212.head(10)
```

```python
# -------------------
# 1. CSV 불러오기
# -------------------
df_original = df_tw_missionPromotion_202212  # 원본 유지

# -------------------
# 2. 번역기 초기화
# -------------------
translator = Translator()

# -------------------
# 3. 한자/중국어 감지 함수
# -------------------
def contains_chinese(text):
    if pd.isna(text):
        return False
    return bool(re.search(r'[\u4e00-\u9fff]', str(text)))

# -------------------
# 4. 번역 함수
# -------------------
def translate_to_korean(text):
    try:
        if pd.isna(text) or not contains_chinese(text):
            return text
        result = translator.translate(str(text), src='zh-cn', dest='ko')
        return result.text
    except Exception as e:
        print(f"⚠️ 번역 실패: {text} -> {e}")
        return text

# -------------------
# 5. 번역 DataFrame 생성
# -------------------
df_translated = df_original.copy()  # 원본 보존

for col in df_translated.columns:
    if df_translated[col].apply(lambda x: contains_chinese(x)).any():
        print(f"🔹 번역 중: {col}")
        df_translated[col] = df_translated[col].apply(translate_to_korean)

# -------------------
# 6. 이제 df_translated로 바로 EDA 가능
# -------------------
print(df_translated.head())
print(df_translated.describe())
```
